replacing data points with null value with mean of that column

```python
from sklearn.model_selection import train_test_split
variable=sd.drop(['Outcome','Pregnancies'],axis=1)
outcome=sd['Outcome']
```

```python
variable.describe()
```

```python
variable.replace(to_replace=0,value=variable.mean(),inplace=True)
```

```python
variable.describe()
```

```python
variable.insert(loc=0,column='Pregnancies',value=sd['Pregnancies'])
```

```python
variable.describe()
```

Splitting of dataset for training and testing (80:20)

```python
variable_train , variable_test , outcome_train , outcome_test = train_test_split(variable,outcome,test_size=0.20,random_state=None)
```

fitting data into randomforestclassifer metod of training machine

```python
from sklearn.ensemble import RandomForestClassifier
machine_model_learner = RandomForestClassifier(max_depth=11,random_state=0)
machine_model_learner.fit(variable_train, outcome_train)
outcome_pred = machine_model_learner.predict(variable_test)
```

checking the hundred minus mean error for Variable

```python
from sklearn.metrics import accuracy_score as mean_error_mius_hundred
mean_error_mius_hundred(outcome_test,outcome_pred) * 100
```

predicting for a new data by taking coustom input


```python
coustom_input = (0,16,72,19,175,25.8,0.587,51)
input_data_as_array = mathematical_operator.asarray(coustom_input)
coustom_input= input_data_as_array.reshape(1,-1)
predict=machine_model_learner.predict(coustom_input)
if predict==1:
  print ("person being tested is diabatic")
else:
  print ("person being tested is not diabatic yet")
```
